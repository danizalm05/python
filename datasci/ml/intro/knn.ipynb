```python
from sklearn import preprocessing
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing
```

```python
#Loading Data car.data
data = pd.read_csv("car.data")
data.head()
```

 The method fit_transform() takes a list (each of our columns) and will return to us an array containing our new values.

```python
le = preprocessing.LabelEncoder()
# translate the labels to integer numbers
buying = le.fit_transform(list(data["buying"]))# [3 3 3 ... 1 1 1].
maint = le.fit_transform(list(data["maint"]))#[3 3 3 ... 1 1 1]
door = le.fit_transform(list(data["door"]))#[0 0 0 ... 3 3 3]
persons = le.fit_transform(list(data["persons"]))#[0 0 0 ... 2 2 2]
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))#[1 2 0 ... 1 2 0]
cls = le.fit_transform(list(data["class"]))#[2 2 2 ... 2 1 3]
```

```python
 cls
```

Recombine our data into a feature list and a label list.
We can use the zip() function to makes things easier.

```python
X = list(zip(buying, maint, door, persons, lug_boot, safety))  # features
y = list(cls)  # labels
print(X[0],'  => ' ,y[0])
print(X[121],'  => ' ,y[121])
 
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.1)
```

Training a KNN Classifier

Creating a KNN Classifier is almost identical to how we created the linear regression model. The only difference is we can specify how many neighbors to look for as the argument n_neighbors.

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=9)
```

```python
model.fit(x_train, y_train)# train our model
```

```python
acc = model.score(x_test, y_test)# score our model
print(acc)
```

<b>Testing Our Model </b>

```python
predicted = model.predict(x_test)
# Create a names list so that we can convert our integer predictions into 
# their string representation 
names = ["unacc", "acc", "good", "vgood"]
#predicted
len(predicted)
```

```python

for x in range(len(predicted)):
 print("Predicted = ", names[predicted[x]], "    Data: ", x_test[x], "Actual: ", names[y_test[x]])
```

<b> Looking at Neighbors</b>

```python
predicted = model.predict(x_test)
names = ["unacc", "acc", "good", "vgood"]

for x in range(len(predicted)):
    print("Predicted: ", names[predicted[x]], "Data: ", x_test[x], "Actual: ", names[y_test[x]])
    # See the neighbors of each point in our testing data
    n = model.kneighbors([x_test[x]], 9, True)
    #This will return to us an array with the index in our data of each neighbor. 
    #If distance=True then it will also return the distance to each neighbor from
    # our data point.
    #x_test[x ] = one  line in data array
    print
    print("[",x,"]")
    print("    N: ", n)
    print("-------------------------")
```
